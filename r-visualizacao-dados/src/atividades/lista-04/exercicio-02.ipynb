# Exercício 2

Importanto a bilbioteca `ggplot2` e lendo a base de dados `pesq_prof.csv`.

```r
suppressPackageStartupMessages(library("ggplot2"))
pesq_prof <- read.csv("pesq_prof.csv")

options(warn=-1) # Desabilita mensagens de warning
```

## Gráficos

```r
ggplot(pesq_prof, aes(x = Área)) + geom_histogram(stat = "count")
```

```r
graf <- ggplot(pesq_prof, aes(x = Área)) + geom_histogram(stat = "count")
graf <- graf + geom_histogram(data = subset(pesq_prof, Área == "Geometria"), stat = "count", 
    fill = "darkblue")
graf
```

```r
ggsave("fig3-1.png", device = "png", dpi = 300)
```

```r
ggplot(pesq_prof, aes(x = Peso, y = Altura)) + geom_point()
```

```r
ggplot(pesq_prof, aes(Peso, Altura, colour = Sexo)) + geom_point()
```

```r
ggplot(pesq_prof, aes(Peso, Altura, colour = Sexo, size = Peso)) + geom_point()
```

```r
graf <- ggplot(pesq_prof, aes(Peso, Altura, colour = Sexo)) + geom_point() + geom_smooth(method = lm)
graf
```

```r
graf + xlab("legenda em x -Peso") + ylab("legenda em y -Altura") + xlim(0, 110) + 
    ylim(0, 200)
```

```r
ggplot(pesq_prof, aes(Peso, Altura, colour = Sexo, size = Peso)) + geom_point() + 
    facet_wrap(~Área)
```
